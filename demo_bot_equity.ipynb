# Демонстрация bot_equity
Ниже два варианта использования: прямой вызов build_equity и обёртка process_log_v2.
Колонка `position_profit` поступает из логов, а `bot_equity` — её накопление.

```python
from pathlib import Path
from src.data.equity_builder import build_equity

candles_path = Path("OHCL/T_GLD_new_tr_1t_ohcl.csv")
log_path = Path("trade_data/T_GLD_new_tr_1t.csv")

equity = build_equity(candles_path, log_path, comis=0.5)
equity[[
    'date_time',
    'action_price',
    'current_pos',
    'position_profit',
    'bot_equity',
    'Equity',
]].head(20)
```

```python
from pathlib import Path
from src.moex_all_in_one_v2 import process_log_v2

result = process_log_v2(
    log_path=Path("trade_data/T_GLD_new_tr_1t.csv"),
    build_equity_report=True,
    show_plot=False,
    plot_column='bot_equity',
    comis=0.5,
)
equity = result.equity
if equity is None:
    raise RuntimeError('Equity не построена — включите build_equity_report или show_plot')
equity[[
    'date_time',
    'current_pos',
    'position_profit',
    'bot_equity',
    'Equity',
]].tail(20)
```
